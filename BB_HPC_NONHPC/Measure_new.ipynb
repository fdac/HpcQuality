```python
#Match y and x
sel<-match(x$repo,y$V2);
x$qc=rep(NA,length(sel));
x$qc[!is.na(sel)] <- y$qc[sel[!is.na(sel)]];

z <- read.table("deltasNoCh.dup",sep=";",header=F);
head(z)
```
