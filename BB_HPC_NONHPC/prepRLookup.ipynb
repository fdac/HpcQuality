```python
#Read prepR.r
prepR <- read.table("prepR.r",sep=';', header=T)
head(prepR)
```

```python
#Read list_all
hpc_repo <- read.table("BB_HPC/list_all",sep=' ',header=F,comment.char="",quote="")
names(hpc_repo) <- c("repoName")
hpc_repo$repoName <- gsub("/", "_", hpc_repo$repoName)
head(hpc_repo)

#Find the repos in prepR.r
rows <- c(match(hpc_repo$repoName,prepR$repo))
allkeywords <- prepR[rows,]
head(allkeywords)
```

```python
row_to_keep <- which(allkeywords$repo != "<NA>")

length(row_to_keep)
allkeywords <- allkeywords[row_to_keep,]
head(allkeywords)
write.table(allkeywords, file = "prepRHPC.r", sep = ";")
```

```python
#read list_keywords
hpc_repo <- read.table("list_keywords",sep=' ',header=F,comment.char="",quote="")
names(hpc_repo) <- c("key","c", "repoName")
hpc_repo

rows <- c(match(hpc_repo$repoName,prepR$repo))
keywords <- prepR[rows,]
keywords
```
